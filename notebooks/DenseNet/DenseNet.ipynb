### Main ideas

The main idea of DenseNet it is to reuse produced feature map. What it means by that?

For example, let's look at traditional neural network:


Assume that we have L layers, and each layer produce feature maps from the previous layers input. And here is some problem - it does not have the state, every time we produce new feature maps, some non-linear transformations.

Let's look at DenseNet block logic (my visualization can be incorrect, but i was trying to visualize it as i could):


How to interpret this image? In the beginning, we have some input with 3 channels, and pass it to the first convolutional layer. Then we have another convolutional layer, to which we pass output of the first layer, but also the input to the block. And you may think that this is like in ResNet - residual connections. Actually yes, but instead of adding - we concatenate the result by channels, and if we look at the output - we see that we pass all inputs of the previous layers and output of the last. So, as a result - we get a huge state that contains all produced feature maps in the block. Such architecture is called DenseBlock. Every input to the layer in DenseBlock consists of the all inputs to the previous layers. 

Such architecture helps:
- reuse feature maps(and as a consequence - reduce the number of params and computational efficiency)
- classify since classifier see whole picture, all feature maps produced by block
- improve flow of information and gradients throughout the network

Above you can see how whole DenseNet is built.

In DenseNet, each DenseBlock layer has k produced feature maps, where k is a growth rate of the network. And in the end, as an output of the DenseBlock we get "collective knowledge" - set of all produced feature maps.
